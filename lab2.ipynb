```python
import numpy as np
```

```python
def rem_outliers(df,column,min_val,max_val):
    col_values= df[column].values
    df[column] = np.where(np.logical_or(col_values<=min_val, col_values>=max_val), np.NaN,col_values)
    return df
```

```python
tstrn = rem_outliers(df= tstrn, column= 'age',min_val = 15, max_val= 90)
tstrn['age'].fillna(-1,inplace= True)
```
